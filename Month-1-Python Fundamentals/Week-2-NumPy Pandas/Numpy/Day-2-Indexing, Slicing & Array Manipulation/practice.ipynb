```python
import numpy as np
```

```python
data=np.arange(1,26).reshape(5,5)

print(data) #original array

print(data[1:4,1:4]) # 3x3 sub array
```

```python
a = np.array([3, -1, 4, -1, 5, -9, 2, -6])

print(a[a>0])

a[a<0]=0

print(a)
```

```python
X = np.random.rand(10, 3)
y = np.random.randint(0,2,10)

np.random.seed(42)
indices = np.random.permutation(len(X))  # shuffled index order
print(indices)  # e.g. [1 4 2 0 3]

X_shuffled = X[indices]
y_shuffled = y[indices]
print(X_shuffled)
print(y_shuffled)
```

```python
a=np.arange(10)

b=a[2:5]
b[0]=100

print(a)

#because here we have used the slicing method which is a view of the original array. So, when we change the value of b, it also changes the value of a. to keep it unchanges we have to use the fancy indexing method.

a=np.arange(10)

b=a[[2,3,4]]
b[0]=100

print(a)
```

```python
scores = np.array([65, 90, 78, 82, 55, 95])

indices=np.argsort(scores)[::-1][:3]

print(indices)

print(scores[indices])
```
