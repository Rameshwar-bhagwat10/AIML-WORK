```python
import numpy as np
import time

# Python list approach
size = 1_000_000
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result = [x + y for x, y in zip(list1, list2)]
print("Python list time:", time.time() - start)

# NumPy approach
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
result = arr1 + arr2
print("NumPy time:", time.time() - start)
```

```python
# Real debugging example
X = np.random.rand(100, 5)   # 100 samples, 5 features
weights = np.random.rand(5, 1)  # 5 features -> 1 output

print(X.shape, weights.shape)  # (100, 5) (5, 1)
predictions = X @ weights       # matrix multiply works because inner dims match (5 == 5)
print(predictions.shape)        # (100, 1)
```

```python
a = np.array([[1, 2], [3, 4]])

f = a.flatten()   # ALWAYS returns a COPY
r = a.ravel()     # returns a VIEW when possible (no data copy → faster)

f[0]=100
print(a) #a IS NOT MODIFIED, because flatten returns a copy
print(f) 
print(r)

r[0] = 99
print(a)  # a IS MODIFIED, because ravel shares memory
# [[99  2]
#  [ 3  4]]

```
