```python
import numpy as np
```

```python
a=np.random.randint(1,100,size=(4,3))
print(a)

print(np.mean(a))
print(np.std(a))

standardize=(a-np.mean(a))/np.std(a)
print(standardize)
```

```python
a=np.random.randint(1,10,size=(3,1))
b=np.random.randint(1,20,size=(1,4))
print(a+b)

#we have studied that if there is 1 we have to not consider it, we have to choose corresponding values, so here we have 1 in a and 1 in b, so we have to choose corresponding values, so we have to choose 3 and 4, so the output will be 3+4=7, similarly for other values.
```

```python
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

x=np.array([1,2,3,4,5,0,1,0])

print(relu_derivative(x))
```

```python
a=np.random.randint(1,10,size=(5,4))

row_sums = a.sum(axis=1, keepdims=True)  # shape (2,1)

normalized = a / row_sums   # (2,3) / (2,1) -> broadcasts correctly!
print(normalized)
```

```python
y_true = np.array([1,1,0,1,0,0,1])
y_pred = np.array([1,0,0,1,0,1,1])


squared_error=(y_true-y_pred)**2
print(squared_error)
mse=np.mean(squared_error)
print(mse)
```
