# Set

```python
s = {1, 2, 3}
print(s)
print(type(s))
```

```python
s = {1,1,2,2,3}
print(s)   # {1,2,3}
```

```python
s = {1,1,2,2,3}

s.add(4)
s.remove(2)
print(s)
```

Set Union

```python
a = {1,2,3}
b = {3,4,5}

print(a | b)
```

Intersection

```python
print(a & b)
```

Difference

```python
print(a-b)
print(b-a)
```

```python
data = [1,2,2,3]
unique = list(set(data)) #remove duplicates
print(unique)
```

```python
features1 = {"age", "income"}
features2 = {"income", "gender"}

common = features1 & features2
print(common)
```

# Dictonary

```python
student = {
    "name":"Ram",
    "age":20
}
print(student)
print(type(student))
```

```python
print(student["name"])
```

```python
student["marks"]=90 #adding new key
student["age"]=21 #update existing key
print(student)
```

```python
student.pop("age") #remove key
```

```python
for key in student:
    print(key, ":", student[key])
```

```python
for k,v in student.items():
    print(k,":",v)
```

```python
data = {
    "student1": {"name": "Ram", "marks": 90},
    "student2": {"name": "Shyam", "marks": 85}
}
print(data)
```

```python
print(student.get("age")) #avoid error if key not found
```

```python
squares={x: x*x for x in range(1,6)}
for k,v in squares.items():
    print(k,":",v)
```
