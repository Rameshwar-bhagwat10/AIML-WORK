# Type Conversion

```python
age=int(input("Enter Your age: "))
print(age)
age+=10
print(age)
```

```python
name=input("Enter Your Name: ")
age=int(input("Enter Your age: "))
print(f"Hello I Am {name}, i am {age} year old, i am studying in SCOE college, i secured 90% in FY")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print(a + b)
```

# Shortcut

```python
a,b=map(int,input("Enter Two Number").split())
a,b=[10,20]
print(a+b)
```

```python
prediction = float(input("Enter model prediction: "))
threshold = 0.5

is_spam = prediction > threshold

print(f"Spam: {is_spam}")
```

```python
marks = float(input("Enter student marks: "))

if marks > 50:
    print("Pass")
else:
    print("Fail")
```
