```python
text="I love AI AI"

word=text.split()

freq={}

for word in word:
    freq[word]=freq.get(word,0)+1

print(freq)
```

```python
bad_words=["stupid","idiot","dumb"]
text="Your are stupid person"

for word in bad_words:
    if word in text:
        text=text.replace(word,len(word)*"*")

print(text)
```
