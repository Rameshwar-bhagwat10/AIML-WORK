# Conditional Statements


```python
marks = 60

if marks > 50:
    print("Pass")
```

```python
marks = 40

if marks > 50:
    print("Pass")
else:
    print("Fail")
    
```

```python
marks = 91

if marks >= 90:
    print("A")
elif marks >= 70:
    print("B")
elif marks >= 50:
    print("C")
else:
    print("Fail")
```

```python
age = 18

if age == 18:
    print("Exactly 18")
```

```python
age = 20
income = 50000

if age > 18 and income > 30000:
    print("Eligible")
```

```python
marks = 60

result = "Pass" if marks > 50 else "Fail"
print(result)
```

```python
prediction = 0.8

if prediction > 0.5:
    print("Spam")
else:
    print("Not Spam")
```

```python
accuracy = 0.87

if accuracy > 0.9:
    print("Excellent model")
elif accuracy > 0.7:
    print("Good model")
else:
    print("Needs improvement")
```
