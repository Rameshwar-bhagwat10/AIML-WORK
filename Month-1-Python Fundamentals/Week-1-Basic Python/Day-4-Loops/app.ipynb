# Loops in Python

```python
for i in range(3):
    print("Hello Ram!")
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(2,20,2):
    print(i)
```

# While Loop

```python
i=1
while i<=5:
    print(i)
    i+=1
```

```python
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
for i in range(10):
    if i==4:
        continue
    print(i)
```

```python
for i in range(5):
    pass
```

```python
for i in range(3):
    for j in range(2):
        print(i,j)
```

```python
for i in range(11):
    if i%2==0:
        print(i)
```

```python
[print(i) for i in range(5)]
```

# Enumerate Function

```python
names = ["Ram", "Shyam"]

for index, name in enumerate(names):
    print(index, name)
```

# Zip Function

```python
a = [1,2,3]
b = [4,5,6]

for x,y in zip(a,b):
    print(x,y)
```

```python
data = [10, 20, 30]

for value in data:
    print(value)
```

```python
for epoch in range(5):
    print(f"Training epoch {epoch}")
```

```python
predictions = [0.2, 0.8, 0.6]

for p in predictions:
    
    if p>0.5:
        print("Positive")
    else:
        print("Negative")
```

```python
data = [10, -5, 20, -2]

for d in data:
    if d > 0:
        print(d)
```
