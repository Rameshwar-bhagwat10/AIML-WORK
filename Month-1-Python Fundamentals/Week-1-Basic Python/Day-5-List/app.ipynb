# List

```python
numbers=[10,20,30,40]
print(numbers)
print(type(numbers))
```

```python
print(numbers[0])
print(numbers[1])
print(numbers[-1])
```

```python
numbers[0]=19
print(numbers)
```

```python
numbers=[1,2,3]
numbers.append(4)
print(numbers)
```

```python
numbers.insert(0,0)
print(numbers)
```

```python
numbers.extend([5,6,7])
print(numbers)
```

```python
numbers.remove(1)   # removes value
numbers.pop()        # removes last
numbers.pop(1)       # removes index wise
print(numbers)
```

```python
list=[10,20,30]
for n in list:
    print(n)
```

```python
for i in range(len(list)):
    print(list[i])
```

```python
print(len(list))
print(max(list))
print(min(list))
print(sum(list))
```

```python
print(list[0:3])
print(list[::2])   # skip by 2
```

```python
square=[x*x for x in range(6)]
print(square)
```

```python
even=[x for x in range(10) if x%2==0]
print(even)
```

```python
a = [1,2,3]
b = a

b.append(4)

print(a)  # [1,2,3,4]
```

```python
b = a.copy()
b.append(5)
print(a)  # [1,2,3,4]
print(b)
```

```python
matrix = [[1,2],[3,4]]

flat = [item for row in matrix for item in row]
print(flat)
```

```python
data = [10, 20, 30, 40]

for d in data:
    print(d)
```

```python
data = [10, -5, 20, -2]

clean = [x for x in data if x > 0]
```

```python
predictions = [0.2, 0.8, 0.6]

results = ["Positive" if p > 0.5 else "Negative" for p in predictions]
print(results)
```

```python
data = [10,20,30]

scaled = [x/10 for x in data]
print(scaled)
```
