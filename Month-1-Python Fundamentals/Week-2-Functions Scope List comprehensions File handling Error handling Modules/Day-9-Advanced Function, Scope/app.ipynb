# Advanced Function And Scope

```python
def func():
    x = 10
    print(x)

func()
```

```python
def func():
    x = 10 #local Scope

func()
print(x)  #gives error
```

```python
x = 100 # Global Scope

def func():
    print(x)

func()
```

```python
x = 10

def func():
    x = 20   # creates new local variable

func()
print(x)   # 10
```

```python
def outer():
    x = 10

    def inner():
        print(x)

    inner()

outer()
```

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()
```

```python
def add_item(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst
```

```python
def stats(a, b):
    return a+b, a-b, a*b

s, d, p = stats(10, 5)
```

```python
def square(x):  #Function Composition
    return x * x

def process(x):
    return square(x) + 10

print(square(5))
print(process(5))
```

```python
def total(*nums):
    return sum(nums)

print(total(1,2,3,4))
```

```python
def info(**data):
    print(data)

info(name="Ram", age=18)
```
