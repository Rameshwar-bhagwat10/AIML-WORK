```python
def check_odd_even(num):
    if num%2==0:
        return "Even"
    else:
        return "Odd"
print(check_odd_even(4))
print(check_odd_even(5))
```

```python
def find_max(a,b,c):
    if a>b:
        if a>c:
            return "A is Greater"
        else:
            return "C is Greater"
    else:
        if b>c:
            return "B is Greater"
        else:
            return "C is Greater"
print(find_max(4,5,6))
print(find_max(7,3,5))
```

```python
def count_vowel(string):
    count=0
    vowel="aeiouAEIOU"
    for char in string:
        if char in vowel:
            count += 1
    return count

print(count_vowel("Rameshwar"))
print(count_vowel("Python Programming"))
```

```python
def rev_string(string):
    return string[::-1]

print(rev_string("Rameshwar"))
print(rev_string("Python Programming"))
```

```python
def sum_list(list):
    total=0
    for x in list:
        total+=x
    return total

print(sum_list([1,2,3,4,5]))
```

```python
def find_positive(list):
    positive_list=[x for x in list if x>0]
    return positive_list

print(find_positive([-2, -1, 0, 1, 2, 3]))
```

```python
def calculator(a,b):
    return a+b,a-b,a*b,a/b

sum,diff,prod,div=calculator(10,5)
print("Sum:",sum,"Difference:",diff,"Product:",prod,"Division:",div)
```
