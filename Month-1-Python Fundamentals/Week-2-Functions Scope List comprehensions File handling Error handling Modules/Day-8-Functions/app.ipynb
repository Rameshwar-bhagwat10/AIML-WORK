```python
nums = [1,2,3]

result = list(map(lambda x: x*2, nums))
result1=[x*2 for x in nums] #second method is list comprehension which is more efficient than map function
print(result)
print(result1)
```

```python
nums = [1,2,3,4]

result = list(filter(lambda x: x % 2 == 0, nums))
result1=[x for x in nums if x % 2 == 0] #second method is list comprehension which is more efficient than filter function
print(result)
print(result1)
```
