# LIST COMPREHENSION


```python
[expression for item in iterable] #Syntax of list comprehension
```

```python
nums = [1,2,3,4]

result = []

for x in nums:
    result.append(x * 2)

print(result)
```

```python
result=[x*2 for x in nums]
print(result)
```

```python
squares=[x*x for x in range(1,6)]
print(squares)
```

```python
result = []

for x in range(10):
    if x % 2 == 0:
        result.append(x)
print(result)
```

```python
result=[x for x in range(10) if x%2==0]
print(result)
```

```python
result=list(map(lambda x:x*2,nums))
print(result)
```

```python
nums = [1,2,3,4]

result = ["Even" if x % 2 == 0 else "Odd" for x in nums]

print(result)
```

```python
matrix=[[1,2],[3,4]]
flat=[]

for row in matrix:
    for item in row:
        flat.append(item)
print(flat)
```

```python
matrix=[[1,2],[3,4]]
flat=[item for row in matrix for item in row]
print(flat)
```

```python
def square(x):
    return x*x

result=[square(x) for x in range(1,6)]
print(result)
```

```python
squares = {x: x*x for x in range(5)}

print(squares)
```

```python
unique = {x for x in [1,2,2,3]}
print(unique)
```
