```python
file=open("data.txt","r")

content=file.read()

print(content)

file.close()
```

```python
file=open("data.txt","r")

content=file.readline()

print(content)
file.close()
```

```python
with open("data.txt","r") as file:
    content=file.read()
    print(content)
```

```python
with open("raw.txt","w") as file:
    file.write("Hello i am Ram")
```

```python
with open("raw.txt","a") as file:
    file.write("\nThis is the new line")
```

```python
with open("raw.txt","r") as file:
    for line in file:
        print(line.strip())
```

```python
with open("num.txt","r") as file:
    total=0

    for line in file:
        total+=int(line.strip())

print(total)
```

```python
with open("data.txt","r") as file:

    for line in file:
        name,marks=line.split()
        print(f"Name: {name}, Marks: {marks}")
```

```python
with open("num.txt", "r") as file:
    data = [int(line.strip()) for line in file]
    print(data)
```

```python
with open("data.txt") as file:
    words = []

    for line in file:
        words.extend(line.lower().split())

print(words)
```

```python
def load_data(file):
    with open(file) as f:
        return [float(x.strip()) for x in f]
    
load_data("num.txt")
```
