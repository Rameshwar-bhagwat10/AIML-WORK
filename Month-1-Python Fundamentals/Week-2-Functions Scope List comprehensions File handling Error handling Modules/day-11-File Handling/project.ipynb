```python
def grade(avg):
    if avg>=80:
        return 'A'
    elif avg>=70:
        return 'B'
    elif avg>=60:
        return 'C'
    else:
        return 'D'

with open("data.txt","r") as file, open("output.txt","w") as out:
    for line in file:

        parts=line.strip().split()

        name=parts[0]

        marks=list(map(int,parts[1:]))

        avg=sum(marks)/len(marks)

        grade_value=grade(avg)

        out.write(f"{name} {avg:.2f} {grade_value}\n")

        print("Avg marks of",name,"is",avg,"and the grade is",grade_value)
```

```python
def calculate_avg(marks):
    return sum(marks) / len(marks)


def get_grade(avg):
    if avg >= 80:
        return "A"
    elif avg >= 60:
        return "B"
    else:
        return "C"


def process_file(input_file, output_file):
    with open(input_file, "r") as file, open(output_file, "w") as out:

        for line in file:
            parts = line.strip().split()

            name = parts[0]
            marks = list(map(int, parts[1:]))

            avg = calculate_avg(marks)
            grade = get_grade(avg)

            out.write(f"{name} {avg:.2f} {grade}\n")


process_file("data.txt", "output.txt")
```
