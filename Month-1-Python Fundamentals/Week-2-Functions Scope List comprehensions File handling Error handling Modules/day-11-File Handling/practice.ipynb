```python
with open("numbers.txt","r") as file:

    total=0
    count=0
    for line in file:
        for num in line.split():
            total += int(num.strip())
            count += 1
avg = total / count if count > 0 else 0
print("Total:",total)
print("Count:", count)
print("Average:", avg)


#shortcut

with open("numbers.txt","r") as file:
    total=[int(num) for line in file for num in line.split()]
print(total)
```
